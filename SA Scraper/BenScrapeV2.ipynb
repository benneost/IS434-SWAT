# Scrapeeeeeeee

```python
import re
import json
import requests
import pandas as pd
import datetime
import time
```

```python
#url link
url = "https://shopee.co.id/HUAWEI-Band-6-Smart-Band-Pemantauan-SpO2-24-Jam-96-mode-olahraga-2-Minggu-Battery-Life-i.39372120.8435235827?sp_atk=7b024d6b-6617-403f-88a3-883e5805d28e&xptdk=7b024d6b-6617-403f-88a3-883e5805d28e"

# https://shopee.co.id/LENOVO-GAMING-3-15-RYZEN-5-5600-8GB-16GB-512SSD-GTX1650-4GB-W11-OHS-15.6FHD-IPS-120HZ-BLK-36ID-i.78730497.10479076174?sp_atk=775b2da8-5648-4975-a9d8-11d50dab585d


r = re.search(r"i\.(\d+)\.(\d+)", url)
shop_id, item_id = r[1], r[2]
ratings_url = "https://shopee.sg/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0"



offset = 0
d = {"username": [], "date": [], "variation": [], "rating": [], "comment": []}
while True:
    data = requests.get(
        ratings_url.format(shop_id=shop_id, item_id=item_id, offset=offset)
    ).json()

    total_reviews = data["data"]["item_rating_summary"]["rating_total"]
    
    # uncomment this to print all data:
    # print(json.dumps(data, indent=4))
    
    i = 1
    for i, rating in enumerate(data["data"]["ratings"], 1):
        d["username"].append(rating["author_username"])
        d["date"].append(datetime.datetime.strptime(time.ctime(rating["ctime"]), "%a %b %d %H:%M:%S %Y"))
        d["variation"].append(rating["product_items"][0]["model_name"])
        d["rating"].append(rating["rating_star"])
        d["comment"].append(rating["comment"])
        

        print(rating["author_username"])
        print(rating["rating_star"])
        print(rating["comment"])
        print("-" * 100)
        

    if i % 20:
        break

    offset += 20
   

df = pd.DataFrame(d)
print(df)
print(total_reviews)
print(i)

```

```python
# uncomment this to save to csv
df.to_csv("data.csv", index=False)
```
