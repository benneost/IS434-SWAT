```python
import string
import pandas as pd
import numpy as np
```

```python
from onemapsg import OneMapClient

Client = OneMapClient("shureenteh@swatmobility.com", "iloveSWAT1209!")
```

```python
df = pd.read_csv('SF Hospital Waypoints Test - Sheet1.csv')
```

```python
df['before_comma'] = df['street'].str.split(",").str[0] 
```

```python
df['after_comma'] = df['street'].str.split(",").str[1]
```

```python
df['afterafter_comma'] = df['street'].str.split(",").str[2]
```

```python
df['without_hospname'] = df['after_comma'] + df['afterafter_comma']
```

```python
df['before_comma'] = df['before_comma'].replace('-','',regex=True)
```

```python
df
```

```python
lat=[]
lon=[]
for index, row in df.iterrows():
    try:
        print(index)
        print(Client.search(row.before_comma))
        lat.append(Client.search(row.before_comma)['results'][0]['LATITUDE'])
        lon.append(Client.search(row.before_comma)['results'][0]['LONGITUDE'])
    except:
        lat.append(0)
        lon.append(0)
```

```python
df['lat']=lat
df['lon']=lon
df
```
