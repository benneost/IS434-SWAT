```python
import warnings 
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#Import file
lscompanies = pd.read_csv('logistics_software_companies.csv')
```

```python
df = lscompanies
df
```

```python
df = df[(df['Operating Status']=="Active")]
df
```

```python
df = pd.DataFrame(df, columns = ['Organization Name' , 'Organization Name URL', 'Headquarters Location'])
df
```

```python
df_headquarters = df['Headquarters Location'].str.split(',', expand=True)
df_headquarters
```

```python
df = pd.concat([df, df_headquarters], axis=1)
df
```

```python
df.rename( columns={0:"State", 1:"City", 2:"Country"} ,inplace=True)
```

```python
df = pd.DataFrame(df, columns = ['Organization Name', 'Organization Name URL' ,'State', 'City', 'Country'])
df
```

```python
df['Country'] = df['Country'].str.strip()
```

```python
test_df = df
```

```python
test_df
```

```python
asian_countries = ['Vietnam', 'Thailand', 'Philippines', 'India', 'China', 'Japan', 'Hong Kong', 'Taiwan', 'Australia']
```

```python
test_df[test_df['Country'].isin(asian_countries)]
```

```python
final_df = test_df[test_df['Country'].isin(asian_countries)]
```
