```python
import warnings 
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#Import file
lscompanies = pd.read_csv('logistics_software_companies.csv')
```

```python
df = lscompanies
df
```

```python
df = df[(df['Operating Status']=="Active")]
df
```

```python
df = pd.DataFrame(df, columns = ['Organization Name' ,'Headquarters Location'])
df
```

```python
df_headquarters = df['Headquarters Location'].str.split(',', expand=True)
df_headquarters
```

```python
df = pd.concat([df['Organization Name'], df_headquarters], axis=1)
df
```

```python
test_df = df
```

```python

test_df = test_df.melt(id_vars=["Organization Name"], 
        var_name="Headquaters Location", 
        value_name="Countries")

test_df
```

```python
test_df = test_df.sort_values(["Organization Name"])
test_df
```

```python
test_df = pd.DataFrame(test_df, columns = ['Organization Name' ,'Countries'])
test_df
```

```python
test_df.isnull().any()
```

```python
test_df.dropna(axis=0, inplace=True) #remove all rows with na's.
test_df.reset_index(drop=True)

test_df.isnull().any()
```

```python
test_df
```

```python
test2_df = test_df
test2_df
```

```python
test2_df['Countries'] = test2_df['Countries'].str.strip()
```

```python
test2_df['Countries'].nunique()
```

```python
test2_df["Values"] = 1
test2_df
```

```python
table = test2_df.pivot_table(index='Countries', columns='Organization Name', values='Values')
table
```

```python
table.to_csv('network.csv', index=True)
```
