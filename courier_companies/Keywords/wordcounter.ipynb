```python
df = pd.read_csv('/Users/shureenteh/Downloads/1smu stuff/year 3 sem 2/is434/shu project data/courier_companies/ID_couriercompanies/jne_ID/jne_allreviews.csv')
df = df.dropna()

df
```

```python
column_titles = ['rating','reviews']
df1 = df.reindex(columns=column_titles)
df1
```

```python
#combining the text into one long string for word cloud 
cloud_string = df1['reviews'].str.cat(sep=' ')
```

```python
# Let's first pull out the top 30 words for each category
words = []
for word in cloud_string.split():
        words.append(word)
Counter(words).most_common()
```

```python
allwords = Counter(words).most_common()
allWordsDf = pd.DataFrame(allwords, columns=['word', 'count'])
allWordsDf
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def sentiment_scores(sentence):
    sid_obj = SentimentIntensityAnalyzer() 
    sentiment_dict = sid_obj.polarity_scores(sentence) 
    if sentiment_dict['compound'] >= 0.05 : 
       return "Positive"
    elif sentiment_dict['compound'] <= - 0.05 : 
        return "Negative"
    else : 
       return "Neutral"
```

```python
allWordsDf['Sentiment'] = pd.DataFrame(allWordsDf['word'].apply(sentiment_scores))
```
